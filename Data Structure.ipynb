### TypeError: unhashable type: 'list'  -> what is this

What is Hashing?

Hashing is a technique Python uses to store and find data very quickly.

Python converts an object into a unique integer called a hash value.

```python
hash('apple')
```

Python uses this hash value like an address/location.

#### Why Hashing is Used

Used for fast lookup in:

* set
* dict

```python
s = {10, 20, 30}
20 in s
```

Python does NOT search one by one. It computes hash(20)

and directly jumps to the location.

That is why sets/dictionaries are very fast.

#### What is Hashable?

An object is hashable if:

1. It has a fixed hash value
2. Its value cannot change

Usually immutable objects are hashable.

```python
print(hash(10))
print(hash("hello"))
print(hash((1,2)))
```

#### What is Unhashable?

Mutable objects are unhashable because their content can change.

hash([1,2])

hash({1,2})

hash({"a":1})

TypeError: unhashable type

#### Why Lists Cannot Be Inside Sets

s = {[1,2,3]}

because list is mutable.

If list changes later:

l.append(4)

**its hash should also change.**
**Then Python would not know where the object is stored internally.**
**This breaks set/dictionary lookup.**

## Important Concept
Set itself can change

```python
s = {1,2,3}
s.add(4)
```

because Python reorganizes the set internally.

**But elements inside set must not change So elements must be hashable/immutable.**

**set is mutable but its element is immutable**

```python
set([1,2,3,4,5,6,5,4])
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)   # union        → {1,2,3,4,5,6}
print(A & B)    # intersection → {3,4}
print(A - B)    # difference   → {1,2}   (in A, not B)
print(A ^ B)    # symmetric    → {1,2,5,6} (not in both)
print(A.issubset(B))   # False
print(A.issuperset(B))  # False
```

### Membership is O(1) — very fast!

```python
# Checking membership
"red" in colors         # True  — O(1) lookup
"red" and "green" in ["red", "green"]  # True  — O(n) lookup (list)

# For large datasets, sets are MUCH faster
```
