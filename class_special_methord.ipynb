### Special Methord

```python
# any mthord inside python which is going to start and end with __ its called as special methord

class course:
    def __init__(self,title,mentor): 
        self.title = title
        self.mentor = mentor

c = course('DS',"GOD")
'''
__init__ --> constructor
it is automatically called when an object of class is created
used to initialize the object attributes
Example: The Course class uses __init__ to take title and mentor as arguments and assign them to the object's attributes (self.title, self.mentor).

constructor is nothing but a kind of special methord which is going to ask me for a data, while creating a object for particular class or any class that we are going to perform
'''
print(c)
```

#### __len__

```python
class course :
    def __init__(self,title,mentor):
        self.title = title
        self.mentor= mentor
        
    def __str__(self):
        return f"my name is sudhanshu {self.title} , {self.mentor}"
    
    def __eq__(self, value):
        return self.title == value.title
    
    def __len__(self):
        return len(self.title) + len(self.mentor)

a = course(["python","data","big data","data science"],"sudhanshu")
```
