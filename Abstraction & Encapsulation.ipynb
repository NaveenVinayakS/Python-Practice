```python
from abc import ABC , abstractmethod
class mentor(ABC): # this is base class - base class means , if a manager wants to write hundreds of function and he likes to distributes among his employees. but he have to create a complete skelton
    @abstractmethod
    def teacher(self): # this is a abstract method
        # there is no implementation in this function who ever is going to call this class will do the implementation
        pass

class student(mentor):

    def teacher(self,subject):
        self.subject = subject
        return f"Teacher is GOD, and he takes {subject}"
    def subjects(self):
        return "I love maths"+self.subject

s = student()
print(s.teacher("maths"))
print(s.subjects())
```

## Encapsulation - public, protected, private

```python
class student:
    def __init__(self,name):
        self.name = name

s = student("naveen")
s.name
```

```python
class student:
    def __init__(self,name):
        self._name = name # protected variable

s1 = student("sudh")
s1._name
```

```python
class student:
    def __init__(self,name):
        self.__name = name # private cannot be used outside the class

s1.student("vinayak")
s1.__name
```

```python
class student:
    def __init__(self,name):
        self.__name = name
        
    def get_name(self):
        return self.__name

s2 = student("Vinayak")
s2.get_name()
```
