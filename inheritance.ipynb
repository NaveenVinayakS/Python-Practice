#### Methord overriding

```python
class mentor_parent:
    def intro(self):
        print("my name is Naveen")
class child(mentor_parent):
    def intro(self):
        print("this is my new child function ")
        return "methord overridden"

ch = child()    

# now which function will be executed ? when intro is called child one or parent one?

ch.intro()

# INTRO() will be executed from child class not from parent class
```

### Super Function

```python
# what if we want to call something intro() inside the parent class also 

class mentor_parent:
    def intro(self):
        print("my name is Naveen")
class child(mentor_parent):
    def intro(self):
        super().intro() # parent class intro() will be executed without getting overridden
        print("this is my new child function ")
        return "methord overridden"

ch = child()    

# now which function will be executed ? when intro is called child one or parent one?

ch.intro()

# INTRO() will be executed from child class not from parent class
```

```python
class course:    
    def __init__(self,title):
        self.title= title
        print(f"parent class init:- {self.title}")

class euroncourse(course):
    def __init__(self,title, mentor):
        super().__init__(title=title)
        self.mentor = mentor
        print(f"child class init:- {self.mentor},{self.title}")

eu = euroncourse("python", "GOD")
```

### Multi-Lable inheritance - child - parent - grand parent

```python
class parent:
    def func1(self):
        return "this is self learning"

class child1(parent):
    def func_child1(self):
        return "this is child1 return"

class child2(child1):
    def func_child2(self):
        return "this is from child2"

c = child2()
print(c.func_child2())
print(c.func_child1())
print(c.func1())
```

### Multiple Inheritance

```python
class CM:
    def edu(self):
        return "education department"

class chief_sec:
    def assembly(self):
        return "assembly"

class PM(CM,chief_sec): # this is multiple inheritance
    pass

p = PM()
print(p.assembly())
print(p.edu())
```
