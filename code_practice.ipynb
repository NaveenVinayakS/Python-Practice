# Given a list of strings, group the anagrams together.

```python
from collections import defaultdict

def group_anagrams(words):
    anagrams = defaultdict(list)
    

    for word in words:
        # Sort the word to create a key
        key = ''.join(sorted(word))
        anagrams[key].append(word)
    print(anagrams)
    return list(anagrams.values())


# Example usage
words = ["eat", "tea", "tan", "ate", "nat", "bat",'nav','vin','niv']

result = group_anagrams(words)

print(result)
```

# longest substring in python

```python
def longest_substring(s: str) -> str:
    seen = {}
    left = 0
    max_len = 0
    start_of_best = 0
    
    for right, char in enumerate(s):
        # If we see a repeat character inside our current window
        if char in seen and seen[char] >= left:
            left = seen[char] + 1
            
        seen[char] = right
        
        # Track the start and length of the longest unique window
        if (right - left + 1) > max_len:
            max_len = right - left + 1
            start_of_best = left
            
    return s[start_of_best : start_of_best + max_len]

s= 'naveen'
longest_substring(s)
```

```python
def longest_substring(s):

    seen = ""
    longest = ""

    for char in s:

        # If duplicate found
        while char in seen:
            print("current char:",char,"current seen:",seen)
            seen = seen[1:]
            print("updated seen:",seen)

        # Add current character
        seen += char

        # Update longest substring
        if len(seen) > len(longest):
            longest = seen
            print("------------------longest:",longest)

    return longest


print(longest_substring("abccabcbb"))
```

```python
def sum_equal(nums,target):
    for i in range(len(nums)):
        for j in range(i+1,len(nums)):
            print(nums[i],"+",nums[j],"=",nums[i]+nums[j])
            if nums[i]+nums[j]==target:
                return i,j
nums = [2, 7, 11, 15]
target = 18

print(sum_equal(nums, target))           
```

## Prime Number Generator

```python
def prime_number(n):
    for i in range(n):
        if i%2!=0:
            yield(i)

for i in prime_number(10):
    print(i)
```

```python
def prime_generator(n):

    for num in range(2, n + 1):
        print("1st loop",num)
        is_prime = True

        for i in range(2, num):
            print("------i=",i,"num=",num)
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            yield num


for prime in prime_generator(10):
    print("prime number=",prime)
```
